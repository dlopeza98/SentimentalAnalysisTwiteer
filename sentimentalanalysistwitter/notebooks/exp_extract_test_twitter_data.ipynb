```python
import os
from kaggle.api.kaggle_api_extended import KaggleApi
import pandas as pd
from dotenv import load_dotenv


# Variables de entorno de Kaggle
KAGGLE_USERNAME = os.getenv("KAGGLE_USERNAME")
KAGGLE_KEY = os.getenv("KAGGLE_KEY")

# Configuración de las credenciales de la API de Kaggle
os.environ["KAGGLE_USERNAME"] = KAGGLE_USERNAME
os.environ["KAGGLE_KEY"] = KAGGLE_KEY

# Autenticar con la API de Kaggle
api = KaggleApi()
api.authenticate()

# Definir y descargar el conjunto de datos
dataset = "goyaladi/twitter-dataset"  # Ajustar
download_path = "../data/01_raw"

if not os.path.exists(download_path):
    os.makedirs(download_path)
api.dataset_download_files(dataset, path=download_path, unzip=True)

# Leer el CSV como un DataFrame
archivo_csv = os.path.join(download_path, "twitter_dataset.csv")  # Ajustar el nombre del archivo
df = pd.read_csv(archivo_csv)

# Mostrar las primeras filas
print(df.head())

# Guardar el DataFrame como CSV
nombre_archivo_guardado = "../data/01_raw/twitter_dataset.csv"
df.to_csv(nombre_archivo_guardado, index=False)
print(f"Dataset guardado como {nombre_archivo_guardado}")
```

```python
from dotenv import load_dotenv
import os
from kaggle.api.kaggle_api_extended import KaggleApi
import pandas as pd

# Cargar variables de entorno
load_dotenv()

# Variables de entorno de Kaggle
KAGGLE_USERNAME = os.getenv("KAGGLE_USERNAME")
KAGGLE_KEY = os.getenv("KAGGLE_KEY")

# Configuración de las credenciales de la API de Kaggle
os.environ["KAGGLE_USERNAME"] = KAGGLE_USERNAME
os.environ["KAGGLE_KEY"] = KAGGLE_KEY

# Autenticar con la API de Kaggle
api = KaggleApi()
api.authenticate()

# Definir y descargar el conjunto de datos
dataset = "goyaladi/twitter-dataset"  # Ajustar
download_path = "./dataset_kaggle"

if not os.path.exists(download_path):
    os.makedirs(download_path)
api.dataset_download_files(dataset, path=download_path, unzip=True)

# Leer el CSV como un DataFrame
archivo_csv = os.path.join(download_path, "twitter_dataset.csv")  # Ajustar el nombre del archivo
df = pd.read_csv(archivo_csv)


# Guardar el DataFrame como CSV
nombre_archivo_guardado = "../data/01_raw/twitter_dataset.csv"
df.to_csv(nombre_archivo_guardado, index=False)
print(f"Dataset guardado como {nombre_archivo_guardado}")
```
